```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('dataset.csv')
df
```

```python
# creating pie chart

Gold_medal=['46','27','26','19','17']
country=['United States', 'Great Britain', 'China', 'Russia', 'Germany']
Color=['r','b','y','g','purple']
plt.pie(Gold_medal,labels=country,colors=Color)
plt.show()
```

```python
# axis equal function

Gold_medal=['46','27','26','19','17']
country=['United States', 'Great Britain', 'China', 'Russia', 'Germany']
Color=['r','b','y','g','purple']
plt.pie(Gold_medal,labels=country,colors=Color)
plt.axis('equal')
plt.show()
```

```python
# using auto percetage function

Gold_medal=['46','27','26','19','17']
country=['United States', 'Great Britain', 'China', 'Russia', 'Germany']
Color=['r','b','y','g','purple']
plt.pie(Gold_medal,labels=country,colors=Color,autopct='%.2f%%')
plt.axis('equal')
plt.show()
```

```python
# using explode function

Gold_medal=['46','27','26','19','17']
country=['United States', 'Great Britain', 'China', 'Russia', 'Germany']
Color=['r','b','y','g','purple']
arr=(0.1,0,0,0,0)
plt.pie(Gold_medal,labels=country,colors=Color,autopct='%.2f%%',explode=arr)
plt.axis('equal')
plt.show()
```

```python
# using shadow function

Gold_medal=['46','27','26','19','17']
country=['United States', 'Great Britain', 'China', 'Russia', 'Germany']
Color=['r','b','y','g','purple']
arr=(0.1,0,0,0,0)
plt.pie(Gold_medal,labels=country,colors=Color,autopct='%.2f%%',explode=arr,shadow=True)
plt.axis('equal')
plt.show()
```

```python
# using legend function

Gold_medal=['46','27','26','19','17']
country=['United States', 'Great Britain', 'China', 'Russia', 'Germany']
Color=['r','b','y','g','purple']
arr=(0.1,0,0,0,0)
plt.pie(Gold_medal,labels=country,colors=Color,autopct='%.2f%%',explode=arr,shadow=True)
plt.legend(country)
plt.axis('equal')
plt.show()
```
