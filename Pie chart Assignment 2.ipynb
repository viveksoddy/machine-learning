```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
```

```python
df=pd.read_csv('computer_parts.csv')
df
```

```python
Bugs= df["Number Of Bugs"]
Product_Name=df["Product Name"]
plt.pie(Bugs,labels=Product_Name )
plt.show()
```

```python
Bugs= df["Number Of Bugs"]
Product_Name=df["Product Name"]
plt.pie(Bugs,labels=Product_Name )
plt.show()
plt.pie(Bugs,labels=Product_Name,autopct='%.f%%',shadow=True)
plt.legend(Product_Name)
plt.axis('equal')
plt.show()
```
